# Counting Cars with a Raspberry Pi and DSP

I recently found myself wanting to count the cars that go past my house throughout the day.  The eventual solution is a Raspberry Pi to listen for cars as they passed.  This has proven surprisingly effective and resilient against the noise throughout the day.  In this post, we'll walk through developing the algorithm in an IPython notebook.  If you just want the final code, it will also be provided.  Note that it has a bunch of tunable parameters, so it's not something you can just use out of the box.  You'll need to know a bit about computers and software to get it running, then you'll need to read the steps below to understand how to "dial it in" for your local counting needs.

The first thing I tried was a webcam with motion detection on, just to see what it was like.  This was a complete bust in the low light at night.  I rent, so I can't attach a new light to the side of my house to help out with this, so the webcams were only getting a few frames per second of video.  Given that cars are only in the frame for a few hundred milliseconds, this meant that we had to detect cars from a single frame of video, which wound up being unusable.

However, while waiting for this to collect data, I noticed that I could tell a car was going by from the sound alone.  It's surprisingly easy on a quiet street: go sit by the window and close your eyes.  You'll be able to hear every car go by, the characteristic "whhhhh" of road noise rising and falling.  This seemed promising.

From a sound-processing point of view, it's a slightly awkward signal to dig out.  It's not just volume: some cars are loud, some are rather quiet.  It's mostly down to how fast they're going and what kind of tires they have.  Interestingly, you can usually hear the quieter ones go by, even when there's lots of other ambient noise.  The thing that allows this is the "whhhhh" sound: it's similar to the sound of a white noise generator (you might have used one of these to get an infant to sleep).  We just need to find that increase in broadband noise.

To test this hypothesis, I recorded a few minutes of audio in my front room, with a mic up next to the window.  Let's go to the spectrogram.  Err, excuse me while I set up the instrumentation here.

```python
# Generic setup to get the data.
%pylab inline
pylab.rcParams["figure.figsize"] = (15,5)

import scipy.io.wavfile
```

```python
Fs, w = scipy.io.wavfile.read("rec_1466238364_300.wav")
specgram(w, Fs=Fs); None
xlabel("Time (seconds)")
ylabel("Frequency")
title("Power in each frequency for each timeslice")
```

What we're looking at here is a graph over time (going left-to-right), with frequencies going from the bottom to the top.  Down at the bottom of the graph is bass, down at 200Hz.  At the top is 4,000 Hz, which is pretty far up the treble (the sound of a tea kettle whistle is around 4,000 Hz).  The colors are kind of like a weather map, showing you the intensity of rain: green is not so much, then yellow, orange, and red is high-intensity.  Each of the big orange-red lumps is a car going past.  

We can see other things in this data as well, which I'd like to show you briefly:

```python
subplot(1,5,1)
specgram(w[(143*Fs):(147*Fs)], Fs=Fs); None
title("Car")

subplot(1,5,2)
specgram(w[(13*Fs):(15*Fs)], Fs=Fs); None
title("Music(?)")

subplot(1,5,3)
specgram(w[(47*Fs):(49*Fs)], Fs=Fs); None
title("Talking")

subplot(1,5,4)
specgram(w[(95*Fs):(97*Fs)], Fs=Fs); None
title("Car door")

subplot(1,5,5)
specgram(w[(284*Fs):(286*Fs)], Fs=Fs); None
title("Car at idle")
```

In the first of these, we see a car roll past.  It's red and orange all over the place, with a bit of yellow and green.

In the second, we see something like music: it has lots of energy from 1500~3000 Hz, with a regular beat to it, but is generally quiet.

In the third graph, we have human speech.  This is characterized by the "harmonics": note that there's a horizontal line at ~600Hz, then another at about twice that, 1200Hz, and a few more in there.  It's similar to the music, except the lines are over time, instead of over frequency.

In the fourth graph, we have a car door slamming shut.  A you can see, it's pretty loud, and seems to have a bit of echo (hence the smudge to the right).  It's very brief, however, and still has a bit more "focus" to its sound.  This is probably the resonance of the car door, making a thumping sound like a drum, instead of a hiss.

The final graph is a car idling somewhere around here.  You can see that it generates a bit of noise, but it's somewhat more concentrated than a car rolling past.  There's a lot of noise, but it's not as evenly distributed as the car noise is.

Based on all of this, an algorithm presents itself.  For each slice of time, we take its spectrum, find the median energy in those buckets, and then track this sequence of median energies.  We don't care what frequency it came from: we're looking for places where there's more white noise.  We smooth this median over a certain span of time, since we know cars have to spend a few seconds in earshot as they go past, and this removes natural variability from our data.  From there, we want to find where the noise is higher than it has been lately: during the day, the city is generally a louder place than it is at night, so what might be aberrantly loud at night is actually normal during the daytime.

Let's actually put this into code, and see how it goes:

```python
def find_cars(Fs,w):
    """Uses some signal heuristics to identify cars going past
    
    This uses an algorithm based on finding local patches of white noise.
    You can tune its behavior using the parameters at the top of the
    function body.  (This is proof-of-concept, not production code.)
    
    Input: 
    Fs: sample rate of wav file
    w: wav file, from scipy.io.wavfile.read()
    
    Returns: (highs, x_d, bins)
    highs: a list of (timestamp, duration) pairs, one per car pass
    x_d: a smoothed version of the z-score of the median bucket power
    bins: timestamps for each x_d entry, to line up with the car passes
    
    """
    averaging_period = 2.0 # seconds to average over
    threshold = 0.3 # threshold for triggering, multiple of std over mean   
    holdoff = 1.0   # minimum seconds to let through
    
    Pxx, freqs, bins = matplotlib.mlab.specgram(w, Fs=Fs, NFFT=128, noverlap=0)

    Fs_d = 1/(bins[1]-bins[0])
    
    # Find the median of each FFT timeslice
    x = np.log(np.percentile(Pxx, 50, axis=0))
    
    # Smooth out by the averaging period
    x_cs = np.cumsum(x)
    N = int(averaging_period/(bins[1]-bins[0]))
    x_d = x_cs[N:] - x_cs[:-N]

    # Find our cutoff value based on stats of body
    cutoff = np.mean(x_d) + threshold * np.std(x_d)

    # Find samples where it's over that threshold
    y = x_d > cutoff
    
    # Jiggery pokery to get down to timestamps
    y_delta = y[1:] != y[:-1]
    y_deltas = y_delta.nonzero()[0]
    y_spans = np.diff(y_deltas)
    
    highs = []
    for k, i in enumerate(y_deltas[:-1]):
        if not y[i] and y_spans[k] >= holdoff*Fs/Fs_d:
            highs.append( [float(i)/Fs_d, y_spans[k]/Fs_d] )
    
    return highs, x_d, bins[:-N]
```

```python
cars, vals, bins = find_cars(Fs,w)
```

```python
plot(bins, vals)

title("Smoothed median power per FFT")
xlabel("Grey bands are identified as car passes")

for ts, l in cars: # highlight car passes
    axvspan(ts, ts+l, color="grey", alpha=0.2)
    axvline(ts+l/2, color="grey", alpha=0.9)
```

```python
specgram(w, Fs=Fs); None
for ts, l in cars:
    axvline(ts+l/2, color="grey")
    
title("Raw FFT, identified cars marked with vertical lines")
None
```

As you can see above, the method described works pretty well.  It misses cars when they come in packs, but so would the human ear.  There's definitely room for improvement, but for my needs, this is good enough.

### A quick defense of this complicated methodology

If you're familiar with DSP, you're probably wondering why I did it this way.  It's very complicated, and for identifying noise, there are probably less computationally-expensive ways to do it.  I don't disagree, but I couldn't find anything that worked better.  Looking at straight volume is hopeless (see below).  Using a high-pass filter, I found it possible for conversations to trigger it when a quiet car pass didn't.  Using low-pass to look for the rumble of the car was laughable: it turns out a major city is full of rumbling noises at all times of day or night.  Who'd've guessed?

For me, this is a tool to get a quick estimate of the traffic on my street over a few weeks. After a month or two, it's going to be pulled down, and I'll probably never use it again.  I thought publishing this might amuse a few people, and a few people have asked for a technical run-down, so, here it is.
